```python
import os
import numpy as np
import pandas as pd
import mne
from scipy.signal import find_peaks
from datetime import datetime, timedelta

# ---------- SETTINGS ----------
EDF_PATHS = [
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaasy\s003_2003\01_tcp_ar\aaaaaasy_s003_t000.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaasy\s003_2003\01_tcp_ar\aaaaaasy_s003_t001.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaasy\s003_2003\01_tcp_ar\aaaaaasy_s003_t002.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaasy\s003_2003\01_tcp_ar\aaaaaasy_s003_t003.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaasy\s003_2003\01_tcp_ar\aaaaaasy_s003_t004.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaasy\s003_2003\01_tcp_ar\aaaaaasy_s003_t005.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaasy\s003_2003\01_tcp_ar\aaaaaasy_s003_t006.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaasy\s003_2003\01_tcp_ar\aaaaaasy_s003_t007.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaasy\s003_2003\01_tcp_ar\aaaaaasy_s003_t008.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaasy\s003_2003\01_tcp_ar\aaaaaasy_s003_t009.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaasy\s003_2003\01_tcp_ar\aaaaaasy_s003_t010.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaasy\s003_2003\01_tcp_ar\aaaaaasy_s003_t011.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaasy\s003_2003\01_tcp_ar\aaaaaasy_s003_t012.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaasy\s003_2003\01_tcp_ar\aaaaaasy_s003_t013.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaasy\s003_2003\01_tcp_ar\aaaaaasy_s003_t014.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaasy\s003_2003\01_tcp_ar\aaaaaasy_s003_t015.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaakm\s007_2003\01_tcp_ar\aaaaaakm_s007_t000.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaakm\s007_2003\01_tcp_ar\aaaaaakm_s007_t001.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaakm\s007_2003\01_tcp_ar\aaaaaakm_s007_t002.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaakm\s007_2003\01_tcp_ar\aaaaaakm_s007_t003.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaakm\s007_2003\01_tcp_ar\aaaaaakm_s007_t004.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaakm\s007_2003\01_tcp_ar\aaaaaakm_s007_t005.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaakm\s008_2003\01_tcp_ar\aaaaaakm_s008_t000.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaakm\s008_2003\01_tcp_ar\aaaaaakm_s008_t001.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaakm\s008_2003\01_tcp_ar\aaaaaakm_s008_t002.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaakm\s008_2003\01_tcp_ar\aaaaaakm_s008_t003.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaakm\s008_2003\01_tcp_ar\aaaaaakm_s008_t004.edf",
    r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\aaaaaakm\s008_2003\01_tcp_ar\aaaaaakm_s008_t005.edf"
    
]

OUTPUT_DIR = r"C:\Users\user\OneDrive\Documents\UCC Studies Sem 1\IS6611_Applied Research in Business Analytics\Dataset\TUH EEG Dataset\OutputsRESP"
os.makedirs(OUTPUT_DIR, exist_ok=True)

def hjorth(sig):
    d1, d2 = np.diff(sig), np.diff(np.diff(sig))
    v0, v1, v2 = np.var(sig), np.var(d1), np.var(d2)
    activity = v0
    mobility = np.sqrt(v1 / v0) if v0 else 0
    complexity = (np.sqrt(v2 / v1) / mobility) if v1 and mobility else 0
    return activity, mobility, complexity

for edf_file_path in EDF_PATHS:
    print(f"\nProcessing {edf_file_path} ...")
    
    base_path = os.path.splitext(edf_file_path)[0]
    annotation_csv_path = base_path + ".csv"
    
    # Load annotations if available
    if os.path.exists(annotation_csv_path):
        ann = pd.read_csv(annotation_csv_path, comment='#')
        ann.columns = ann.columns.str.strip()
        seizures = ann[~ann['label'].str.contains('bckg', case=False)].sort_values(by='start_time')
        seizure_intervals = seizures[['start_time', 'stop_time']].values if not seizures.empty else np.array([])
    else:
        seizure_intervals = np.array([])
        print(f"Warning: No annotation CSV found for {edf_file_path}. Seizure labels set to 0 and seizure_in_seconds to NaN.")
    
    raw = mne.io.read_raw_edf(edf_file_path, preload=True, verbose=False)
    sfreq = int(raw.info['sfreq'])
    
    eeg = [c for c in raw.ch_names if 'EEG' in c.upper()]
    emg = [c for c in raw.ch_names if 'EMG' in c.upper()]
    ekg = [c for c in raw.ch_names if 'EKG' in c.upper()]
    resp = [c for c in raw.ch_names if 'RESP' in c.upper()]
    selected_channels = list(dict.fromkeys(eeg + emg + ekg + resp))
    
    if eeg:
        raw.set_eeg_reference('average', projection=True)
    raw.filter(0.5, 40.0, verbose=False)
    
    target_fs = 250
    if sfreq != target_fs:
        raw.resample(target_fs, npad="auto", verbose=False)
        sfreq = target_fs
    
    data, times = raw.get_data(picks=selected_channels, return_times=True)
    
    patient_id = os.path.splitext(os.path.basename(edf_file_path))[0]
    
    custom_date = datetime(2025, 7, 2, 0, 0, 0)
    
    win_s, step_s = 10.0, 5.0
    n_win = int(win_s * sfreq)
    n_step = int(step_s * sfreq)
    
    rows = []
    
    for start_idx in range(0, len(times) - n_win + 1, n_step):
        end_idx = start_idx + n_win
        window_data = data[:, start_idx:end_idx]
        t0, t1 = times[start_idx], times[end_idx - 1]
        
        # ---------- Seizure annotation logic ----------
        label_num = 0
        seizure_in_seconds = np.nan
        
        overlapping_seizure = False
        next_seizure_start = None
        last_seizure_end = None
        
        for s_start, s_stop in seizure_intervals:
            if (t1 >= s_start) and (t0 <= s_stop):
                overlapping_seizure = True
                break
            if s_start > t1:
                if next_seizure_start is None or s_start < next_seizure_start:
                    next_seizure_start = s_start
            if s_stop < t0:
                if last_seizure_end is None or s_stop > last_seizure_end:
                    last_seizure_end = s_stop
        
        if overlapping_seizure:
            label_num = 1
            seizure_in_seconds = 0
        else:
            if next_seizure_start is not None and next_seizure_start > t1:
                seizure_in_seconds = round(next_seizure_start - t1, 2)
            elif last_seizure_end is not None and last_seizure_end < t0:
                seizure_in_seconds = round(-(t0 - last_seizure_end), 2)
            else:
                seizure_in_seconds = 0
        # ----------------------------------------------
        
        feat = []
        for sig in window_data:
            mean_val = sig.mean()
            std_val = sig.std()
            energy = np.dot(sig, sig)
            max_val = sig.max()
            min_val = sig.min()
            activity, mobility, complexity = hjorth(sig)
            feat += [mean_val, std_val, energy, max_val, min_val, activity, mobility, complexity]
        
        # Heart rate from EKG
        if ekg:
            ekg_signal = window_data[selected_channels.index(ekg[0])]
            norm_ekg = (ekg_signal - np.mean(ekg_signal)) / np.std(ekg_signal)
            peaks, _ = find_peaks(norm_ekg, distance=int(sfreq * 0.4))
            if len(peaks) > 1:
                rr_intervals = np.diff(peaks) / sfreq
                avg_rr = np.mean(rr_intervals)
                heart_rate = 60.0 / avg_rr
            else:
                heart_rate = np.nan
        else:
            heart_rate = np.nan
        
        bpm = np.nan
        resp_channel = None
        if 'EEG RESP1-REF' in selected_channels:
            resp_channel = 'EEG RESP1-REF'
        elif 'EEG RESP2-REF' in selected_channels:
            resp_channel = 'EEG RESP2-REF'
        
        if resp_channel:
            resp_signal = window_data[selected_channels.index(resp_channel)]
            min_breath_interval = int(sfreq * 1.5)
            resp_peaks, _ = find_peaks(resp_signal, distance=min_breath_interval)
            bpm = len(resp_peaks) * (60 / win_s)
        
        window_time = custom_date + timedelta(seconds=t0)
        time_str = window_time.strftime("%d-%m-%Y %H:%M:%S")
        
        feat += [time_str, heart_rate, label_num, seizure_in_seconds, patient_id, bpm]
        rows.append(feat)
    
    header = []
    for ch in selected_channels:
        header += [f"{ch}_mean", f"{ch}_std", f"{ch}_energy", f"{ch}_max", f"{ch}_min",
                   f"{ch}_hjorth_activity", f"{ch}_hjorth_mobility", f"{ch}_hjorth_complexity"]
    header += ["Time", "Heart_Rate", "label", "seizure_in_seconds", "patient_id", "Resp_BPM"]
    
    output_file = os.path.join(OUTPUT_DIR, f"features_{patient_id}.csv")
    df = pd.DataFrame(rows, columns=header)
    df.to_csv(output_file, index=False)
    print(f"Saved features to {output_file}")
```
